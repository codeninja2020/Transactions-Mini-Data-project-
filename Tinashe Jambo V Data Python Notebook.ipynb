```python
import pandas as pd
import numpy as np
import re
```

```python
# TODO 1 importing CSV
transactionsdf = pd.read_csv('transaction.csv',sep=';')
#preview data Dataframe
transactionsdf.tail(3)
```

```python
#TODO 1 DESCRIBE 
transactionsdf.describe().apply(lambda s: s.apply('{0:.2f}'.format))
```

```python
#TODO 2
transactionsdf['CostPerTransaction']=transactionsdf['NumberOfItemsPurchased']*transactionsdf['CostPerItem']
# Show head 
transactionsdf.head(3)
```

```python
#TODO 3 
transactionsdf['SalesPerTransaction']=transactionsdf['NumberOfItemsPurchased']*transactionsdf['SellingPricePerItem']
transactionsdf.head(3)
```

```python
# TODO 4
transactionsdf['ProfitPerTransaction']=transactionsdf['SalesPerTransaction']-transactionsdf['CostPerTransaction']
#show head
transactionsdf.head(3)
```

```python
#TODO 5
transactionsdf['ProfitPerTransaction'] = transactionsdf['ProfitPerTransaction'].round().astype(int)
# show head
transactionsdf.head(3)
```

```python
#TODO 6 
# In the transaction.csv file, the dates are split into 3 separate columns: Day, Month
# and Year. Combine these 3 columns to create a new date column in the format
# of “day-month-year” i.e 2-July-2019

cols = ["Day","Month","Year"]
transactionsdf['Date'] = transactionsdf[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
#show head
transactionsdf.head(3)
```

```python
# TODO 7
# Change the ItemDescription column in the dataframe to lowercase
transactionsdf['ItemDescription'] = transactionsdf['ItemDescription'].str.lower()
# Show Head
transactionsdf.head(2)
```

```python
# TODO 8
reg=r"'([^']*)'" #single qoutes
newList = []
for row in transactionsdf['ClientKeywords']:
    newList.append(re.findall(reg, row))
```

```python
# Create a new dataframe 
newDf = pd.DataFrame(newList,columns=['ClientAge', 'ClientType', 'LengthofContract'])
newDf.head(3)
```

```python
#Add new columns
transactionsdf= transactionsdf.assign(ClientAge=newDf['ClientAge'],ClientType=newDf['ClientType'],LengthofContract=newDf['LengthofContract'])

#TODO 9 
transactionsdf=transactionsdf.drop(['ClientKeywords'],axis=1)
#preview dataframe 
transactionsdf.tail(3)
```

```python
#SAVE to CSV
transactionsdf.to_csv('TransactionsSubmission.csv')
```
